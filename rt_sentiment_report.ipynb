# Rotten Tomatoes Sentiment Classification

## Research Question

Can short movie reviews be classified as positive or negative using TF-IDF features and standard machine learning models?



## Data

The dataset consists of Rotten Tomatoes movie reviews labeled as positive (1) or negative (0). Each entry includes the review text and a binary sentiment label.

There are about 10,000 reviews, and the dataset is roughly balanced between the two classes.

Source:
https://www.kaggle.com/datasets/thedevastator/movie-review-data-set-from-rotten-tomatoes




## Methods

The review text was converted into numerical features using 'TfidfVectorizer'.

To avoid data leakage, TF-IDF transformation and model training were combined using a scikit-learn `Pipeline`, so preprocessing occurred within each fold during cross-validation.

The following models were compared:

- Logistic Regression  
- Linear SVM  
- XGBoost  

Hyperparameters were tuned using `GridSearchCV` with 5-fold cross-validation. Model performance was evaluated using F1-score and accuracy on the test set.


## Results

The best classical model achieved:

- Accuracy $\approx$ 0.77  
- F1 $\approx$ 0.77  

Logistic Regression and Linear SVM performed similarly, while XGBoost performed slightly worse on this dataset.


## Conclusion

TF-IDF features combined with linear models work reasonably well for short sentiment classification tasks. For this dataset, more complex models did not provide a clear advantage over simpler linear approaches.
