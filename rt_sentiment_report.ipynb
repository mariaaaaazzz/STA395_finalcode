# Rotten Tomatoes Sentiment Classification

## Research Question

Can short movie reviews be classified as positive or negative using TF-IDF features and classical machine learning models?



## Data

The dataset consists of Rotten Tomatoes movie reviews labeled as either positive (1) or negative (0). Each observation includes the review text and a binary sentiment label.  

The dataset contains about 10,000 reviews and is balanced between the two classes.

Source:  
https://www.kaggle.com/datasets/thedevastator/movie-review-data-set-from-rotten-tomatoes

An 80/20 train-test split was used with `random_state=42`.



## Methods

Text data was converted into numerical features using `TfidfVectorizer`.

A scikit-learn `Pipeline` was used to combine TF-IDF feature extraction and model training. This ensures that preprocessing occurs inside each fold during cross-validation and avoids data leakage.

Three models were evaluated:

- Logistic Regression  
- Linear Support Vector Machine  
- XGBoost  

Hyperparameters were tuned using `GridSearchCV` with 5-fold cross-validation.  
F1-score was used as the primary selection metric, along with accuracy for reporting.



## Results

The best-performing classical model achieved:

- Accuracy ≈ 0.77  
- F1 ≈ 0.77  

Logistic Regression and Linear SVM performed similarly.  
XGBoost performed slightly worse on this dataset.



## Conclusion

TF-IDF based linear models provide reasonable performance for binary sentiment classification on short reviews. For this dataset, simpler linear models were competitive and stable after cross-validation tuning.
